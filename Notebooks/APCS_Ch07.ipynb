## Topic 1-函式定義 function definition and call

### Step 1: Calculate the greatest common factor gcd of 240 and 96

```python
a=240; b=96
r=a%b
while r!=0:
  a=b;b=r
  r=a%b
print(f"gcd={b}")
```

### Step 2: Define a function gcd()

* parameter參數：函數定義中的型式變數 def gcd(x, y):
* argument引數：函數呼叫時的真實變數  gcd(a,b)

```python
def gcd(x, y):
  r=x%y
  while r!=0:
    x=y;y=r
    r=x%y
  return y
```

```python
a=240; b=96
print(f"{a}, {b}, {gcd(a,b)}")
```

### Step 3: 階乘 (factorial)

```python
n = 5
v = 1
for i in range(1, n + 1):
    v *= i
print(v)
```

### Step 4: Define a function factorial()

```python
def factorial(n):
    v = 1
    for i in range(1,n+1):
        v *= i
    return v

print(factorial(5))
```

### Step 7: return value

```python
#------------------------------------
# 從1加到100
#------------------------------------
def add100():
    i = 1
    sum = 0
    while i <= 100:
        sum += i
        i += 1
    return sum

a = add100()
print(a)
print(type(a))
```

### Step 8: multiple return values

```python
def cm_2_feet_and_inch(cm):
    inch = cm/2.54
    feet = inch //12
    inch = inch %12
    return feet, inch
```

## Topic 3-遞迴函數 (Recursion)

### Step 1: 使用迴圈的階乘 (Factorial)計算

$n! = 1 * 2 * ... * n$

```python
def factorial_loop (n):
    factor = 1
    for i in range(1,n+1):
        factor *= i
    return factor
print(factorial_loop(10))
```

### Step 2: 使用遞迴的階乘 (Factorial)計算

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial_loop(10))
```

```python
#IPO-input
n,m=map(int,input().split())
mat=[]
for i in range(n,m+1):
  vv=list(map(int,input().split()))
  mat.append(vv)

dd=dict()
for i in range(n):
  for j in range(m):
    dd[(i,j)]=mat[i][j]

#IPO-process
pp=[]
for i in range(n):
  for j in range(m):
    x=mat[i][j]
    s=0
    for a,b in dd:#離(i,j)曼哈頓距離為x內的點數值總和
      if abs(a-i)+abs(b-j)<x:
        s+=dd[(a,b)]
    if s%10==x:#點數值總和%10恰為x
      pp.append((i,j))

#IPO-output
print(len(pp))
for i,j in pp:
  print(i,j)
```
